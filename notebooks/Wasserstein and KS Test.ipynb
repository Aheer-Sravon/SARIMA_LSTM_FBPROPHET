# Kolmogorov-Smirnov Test

## SARIMA

```python
# Store 1
print("\nPerforming KS Test for Store 1 - \n")

df1 = pd.read_csv("../scripts/log/walmart_sarima_actual_vs_predicted_1.csv")

# Perform KS test
ks_statistic, ks_pvalue = ks_2samp(df1['Actual'], df1['Predicted'])

print(f"KS Statistic: {ks_statistic:.4f}")
print(f"P-value: {ks_pvalue:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if ks_pvalue < alpha:
    print(f"Result: Reject null hypothesis (p < {alpha})")
    print("The two distributions are significantly different.")
else:
    print(f"Result: Fail to reject null hypothesis (p ≥ {alpha})")
    print("No significant difference between distributions.")


print("\nPerforming KS Test for Store 3 - \n")

df3 = pd.read_csv("../scripts/log/walmart_sarima_actual_vs_predicted_3.csv")

# Perform KS test
ks_statistic, ks_pvalue = ks_2samp(df3['Actual'], df3['Predicted'])

print(f"KS Statistic: {ks_statistic:.4f}")
print(f"P-value: {ks_pvalue:.4f}")

# Interpretation
alpha = 0.05  # Significance level
if ks_pvalue < alpha:
    print(f"Result: Reject null hypothesis (p < {alpha})")
    print("The two distributions are significantly different.")
else:
    print(f"Result: Fail to reject null hypothesis (p ≥ {alpha})")
    print("No significant difference between distributions.")
```
