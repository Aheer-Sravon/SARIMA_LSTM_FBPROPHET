```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
df = pd.read_excel("Summary.xlsx", sheet_name="Sheet1")

# Drop irrelevant columns
df = df[['Name of the method', 'MAPE', 'MAE', 'RMSE']].dropna()

# Rename 'MAE' to 'MSE' for consistency in plotting
#df.rename(columns={'MAE': 'MSE'}, inplace=True)

# Set plot style
sns.set_style("whitegrid")

# Define function to plot bar charts
def plot_bar_chart(metric, title, color):
    plt.figure(figsize=(8, 5))
    sns.barplot(x='Name of the method', y=metric, data=df, palette=color)
    plt.xlabel("Forecasting Methods", fontsize=12)
    plt.ylabel(metric, fontsize=12)
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xticks(rotation=30, ha='right', fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()


plot_bar_chart('MAPE', 'Mean Absolute Percentage Error (MAPE)', "Blues")


```

```python
plot_bar_chart('MAE', 'Mean Absolute Error (MAE)', "Reds")
```

```python
plot_bar_chart('RMSE', 'Root Mean Squared Error (RMSE)', "Greens")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the relevant portion of the dataset
data = {
    "DATE": ["2/1/2025", "2/2/2025", "2/3/2025", "2/4/2025", "2/5/2025", "2/6/2025", "2/7/2025"],
    "ACTUAL AMOUNT": [11, 7, 25, 12, 24, 12, 16],
    "SARIMA": [7.19, 8.01, 8.00, 7.98, 7.74, 6.86, 6.27],
    "SARIMAX": [6.96, 7.31, 7.77, 9.72, 11.00, 12.31, 12.28],
    
}

df = pd.DataFrame(data)

# Convert DATE to datetime format
df['DATE'] = pd.to_datetime(df['DATE'])

# Set plot style
sns.set_style("whitegrid")

# Plot time-series comparison
def plot_time_series():
    plt.figure(figsize=(10, 6))
    plt.plot(df['DATE'], df['ACTUAL AMOUNT'], marker='o', linestyle='-', label='Actual', color='black')
    plt.plot(df['DATE'], df['SARIMA'], marker='s', linestyle='--', label='SARIMA', color='blue')
    plt.plot(df['DATE'], df['SARIMAX'], marker='^', linestyle='--', label='SARIMAX', color='purple')
    
    plt.xlabel("Date", fontsize=12)
    plt.ylabel("Amount Sold", fontsize=12)
    plt.title("Actual vs Forecasted Values", fontsize=14, fontweight='bold')
    plt.xticks(rotation=30, ha='right', fontsize=10)
    plt.legend()
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Generate time-series plot
plot_time_series()
```
